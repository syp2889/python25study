```python
# 02_01의 훈련세트와 테스트세트를 나누어 평가를 진행해 본 결과 문제가 발생했네요.
# 길이가 25cm, 무게가 150g이면 사람이 판단했을 때는 도미인데, AI는 빙어로 결론이 나왔어요.
# 이 원인을 찾아보고 개선해봅시다.
```

```python
# 도미(35개)와 빙어(14개) 데이터 49개
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]

fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

fish_data = [[l,w] for l, w in zip(fish_length, fish_weight)]   # 2차원 배열로 만듬
fish_target = [1]*35 + [0]*14        # 앞에서 35개는 1, 뒤에 14개는 0
```

```python
# 전에는 파이썬 리스트를 for문과 zip을 이용해서 배열로 만들었는데
# 넘파이에는 이 기능이 내장되어 있는 메서드가 있어요.
import numpy as np # 넘파이 배열용 파이썬 라이브러리
np.column_stack(([1,2,3], [4,5,6])) #.column(열)_stack(쌓기)(([길이용],[무게용]))>>변환
#array([[길이1,무게1],
#       [길이2,무게2],
#       [길이3,무게3]])
```

```python
fish_data = np.column_stack((fish_length,fish_weight))
# fish_data 변수에 2차원 배열을 넣는다 (길이,무게)
print(fish_data[:5]) # 앞쪽 5개를 미리보기 해볼게요. (0~4번 인덱스)
```

```python
# target이 1이면 도미, 0이면 빙어의 정답을 만들기 [1]*35+[0]*14
# 정답파일을 생성해볼게요
print(np.ones(5)) # 1로 채우기 테스트
```

```python
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
# .concatenate() = 첫번째 차원을 따라 배열을 연결합니다.
# 1차원배열 (1이 35개: 도미, 0이 14개: 빙어)
print(fish_target)
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, stratify=fish_target, random_state=42)

# https://wikidocs.net/193722
# train_test_split은 앞으로도 많이 쓸거니 알아두세여
print(train_input.shape, test_input.shape)

# train_input 훈련용 입력값
# test_input 테스트용 입력값
# train_target 훈련용 정답
# test_target 테스트용 정답
# stratify(균일하게) : 타겟값을 넣어줌(target의 class비율을 유지한 채 데이터셋을 split함)
# random_state : 시드값을 42로 지정하고 랜덤 적용(수업용)

#결과: 랜덤하게 처리하면서 훈련용(36객체, 2특성) 테스트용(13객체, 2특성)
```

```python
print(train_target.shape, test_target.shape) # 결과가 튜플로 나옴
```

```python
print(train_target) # 훈련용 정답 데이터
```

```python
print(test_target) # 테스트용 검증 데이터 >> 결론 : 적절하게 분배됨

# 전처리 데이터는 여기서 끝!
```

```python
from sklearn.neighbors import KNeighborsClassifier # k-이웃 근접 알고리즘

kn = KNeighborsClassifier() # 객체 생성
kn.fit(train_input, train_target) # 훈련시작! (트레이닝입력, 트레이닝정답)
```

```python
kn.score(test_input, test_target) # 평가시작! (테스트입력, 테스트정답)
```

```python
print(kn.predict([[25,150]])) # 여기도 마찬가지로 0: 빙어로 나와요.
# 어제 수업한거랑 동일한거라서.. 여기선 split이랑 랜덤섞기를 메서드로 처리하는 법을 배웠어요.
```

```python
# 그래프로 확인해볼게요.
import matplotlib.pyplot as plt

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25,150, marker='^')
# 마커 종류 변경방법은 여기서 확인(파이썬 시각화) https://pyvisuall.tistory.com/39
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 결론: x와 y 축의 비율이 다르기에 정확한 그래프로 판단할 수 없어요. (x는 40까지지만 y는 1000까지있기 때문에 보이는 그래프랑 실제랑 다름)
# k-최근접 이웃 알고리즘은 주변에 있는 5개를 보기 때문에 좀 더 알아볼 필요가 있네요

distances, indexes = kn.kneighbors([[25,150]])
# 내가 알고싶은 값의 거리와 인덱스를 알아볼게요
plt.scatter(train_input[:,0], train_input[:,1]) # 입력한 데이터의 값
plt.scatter(25,150, marker='^')                 # 알고 싶은 값
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D') # 이웃의 값
plt.xlim(0,1000)      # x축에 범위를 0~1000까지 지정(weight)
plt.ylim(0,1000)      # y축에 범위를 0~1000까지 지정
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
# 결론: 녹색으로 25,150에 대한 이웃이 누군지 파악됐네요
# 축의 값범위의 위험성을 알고가자!
```

```python
print(train_input[indexes]) # k-최근접 알고리즘에 근접한 5개의 값 (도미1, 방어4)
#[[ 25.4 242.]>>도미
# [ 15.  19.9]>>빙어
# [ 14.3 19.7]>>빙어
# [ 13.  12.2]>>빙어
# [ 12.2 12.2]>>빙어] >>그래서 빙어라고 판단하고 0으로 나온거네요..
```

```python
print(train_target[indexes]) # 도미1 방어4
```

```python
print(distances) #이웃간의 거리 출력
```

```python
# 두 특성(길이, 무게)의 값이 놓은 범위가 매우 다름 >> 두특성의 스케일이 다르다고 해요.
# 특성간의 스케일이 다른 일은 흔히 발생합니다. (결론: 스케일을 맞춰야 함)
# 특성값을 일정한 기준으로 맞춰주어야 해요. (데이터 전처리라고 함)

# 평균 구해보기 (가장 널리 사용하는 전처리 방법은: 표준점수, z점수라고 함)
# 표준점수는 각 특성값이 평균에서 표준편차의 몇 배만큼 떨어져 있는지를 나타내요.
# 분산은 데이터에서 평균을 뺀 값을 모두 제곱한 다음 평균을 내어 구합니다.
# 표준편차는 분산의 제곱근으로 데이터가 분산된 정도를 나타내요.
# 표준점수는 각 데이터가 원점에서 몇 표준편차만큼 떨어져 있는지를 나타는 값이에요.

mean=np.mean(train_input, axis=0)  # 평균 구하기 axis: 중심선
std= np.std(train_input, axis=0)    # 표준편차 구하기

# train_input >> 36개의 요소와 2개의 특성을 가져요 (36,2)크기의 배열입니다.
# 행을 따라 각 열의 통계 값을 계산해요.
print(mean, std)
# [27.29722222 454.09722222][9.98244253 323.29893931]
# [길이의 평균  무게의 평균][길이의 편차 무게의 편차]
```

```python
# 원본 데이터에서 평균을 빼고 표준편차로 나누어 표준점수를 변환할게요
train_scaled = (train_input - mean) / std # 표준점수 공식 (암기필수!)
# 원본데이터에서 평균을 빼고 표준편차로 나눕니다.
# 넘파이는 train_input의 모든 행에서 mean에 있는 두 평균값을 빼주는거에요.
# std에 있는 두 표준편차를 다시모든 행에 적용합니다.(브로드캐스팅)
```

```python
# 위 공식대로 적용을 하면 41의 길이와 975의 무게의 스케일이 1.37, 1.61로 표준화돼요.

# 전처리 데이터를 이용해서 모델 훈련을 시켜봅시다.

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.xlabel("scaled_length")
plt.ylabel("scaled_weight")
plt.show()
# 차트의 범위가 x축은 -2.0, -1.5, y축은 -1.5~1.5 사이로 비슷해져요.(스케일 조정성공)
# 전에 만들었던 표준점수 이전 차트랑 비교하면 비슷한 차트가 보입니다.
```

```python
# 전처리 데이터를 이용해서 모델 훈련 시키기

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(25,150,marker='^')  # 알고 싶은 추가 데이터
plt.xlabel("length")
plt.ylabel("weight")
plt.show() # 오른쪽 맨 꼭대기에 수상한 샘플이 생겼어요
# 전처리된 데이터는 표준점수를 위해서 값이 축소 되어 있는데
# 알고자 하는 값은 표준점수화가 안된거예요.
# 훈련세트를 평균으로 빼고 표준편차로 나누어 주었기 때문에 값의 범위가 달라진겁니다.
```

```python
# 동일한 기준으로 샘플 변환하고 출력하기
new= ([25,150]-mean)/std
print(f"표준점수로 변환된 25,150: {new}")

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.xlabel("scaled_length")
plt.ylabel("scaled_weight")
plt.show()
# x, y축의 범위가 비슷해요. (-1.5~1.5사이의 값)
```

```python
# 훈련 시작!

kn.fit(train_scaled, train_target) # 스케일 조정된 입력값으로 타겟을 훈련
```

```python
# 테스트 진행 전 표준점수로 변환해야 합니다.

test_scaled = (test_input - mean) / std # 표준점수화
```

```python
kn.score=(test_scaled, test_target) # 평가 진행

new=([25,150]-mean)/std
print(kn.predict([new]))  #25,150짜리 결과를 보면
# [1.]도미로 예측 성공
```

```python
# kneighbors()로 최근접 이웃을 구한 다음 산점도로 그려봅시다.
distances, indexes=kn.kneighbors([new])
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.scatter(train_scaled[indexes,0], train_scaled[indexes,1], marker='D')
plt.xlabel("scaled_length")
plt.ylabel("scaled_weight")
plt.show()
```
